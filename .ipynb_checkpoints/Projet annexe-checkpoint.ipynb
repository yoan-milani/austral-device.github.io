```python
import pandas as pd

class PVRecommender:
    def __init__(self):
        self.data = {
            'Surface': [],
            'Consommation': [],
            'Budget': [],
            'Recommandation': []
        }

    def get_user_input(self):
        surface = float(input("Entrez la surface disponible (m²): "))
        consommation = float(input("Entrez la consommation énergétique (kWh/mois): "))
        budget = float(input("Entrez votre budget (€): "))

        return surface, consommation, budget

    def recommend_pv(self, surface, consommation, budget):
        if surface < 10 and consommation < 200:
            return "Panneaux PV de petite taille (5 kW)"
        elif surface < 20 and consommation < 500:
            return "Panneaux PV de taille moyenne (10 kW)"
        else:
            return "Panneaux PV de grande taille (20 kW)"

    def run(self):
        surface, consommation, budget = self.get_user_input()
        recommendation = self.recommend_pv(surface, consommation, budget)
        print(f"Recommandation : {recommendation}")

if __name__ == "__main__":
    app = PVRecommender()
    app.run()
```

```python
import tkinter as tk
from tkinter import messagebox

class PVRecommender:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("Recommandateur de Centrales PV")

        # Création des champs de saisie
        self.surface_label = tk.Label(self.window, text="Surface Disponible (m²):")
        self.surface_label.pack()
        self.surface_entry = tk.Entry(self.window)
        self.surface_entry.pack()

        self.consommation_label = tk.Label(self.window, text="Consommation Énergétique (kWh/mois):")
        self.consommation_label.pack()
        self.consommation_entry = tk.Entry(self.window)
        self.consommation_entry.pack()

        self.budget_label = tk.Label(self.window, text="Budget (€):")
        self.budget_label.pack()
        self.budget_entry = tk.Entry(self.window)
        self.budget_entry.pack()

        # Bouton pour lancer la recommandation
        self.recommend_button = tk.Button(self.window, text="Recommander", command=self.recommend_pv)
        self.recommend_button.pack()

        # Zone pour afficher la recommandation
        self.recommendation_label = tk.Label(self.window, text="Recommandation:")
        self.recommendation_label.pack()
        self.recommendation_text = tk.Text(self.window, height=5, width=40)
        self.recommendation_text.pack()

    def recommend_pv(self):
        try:
            surface = float(self.surface_entry.get())
            consommation = float(self.consommation_entry.get())
            budget = float(self.budget_entry.get())

            # Hypothèses : 
            # - Puissance nominale par panneau : 300 W
            # - Efficacité de conversion : 15%
            # - Nombre d'heures d'ensoleillement par jour : 5 heures (moyenne)
            # - Jours d'ensoleillement par mois : 30 jours
            # - Dimensions des panneaux : 1.6 m x 1 m

            # Calcul de la puissance requise
            puissance_requise = consommation / (5 * 30 * 0.15)  # kWh/mois / (heures/jour * jours/mois * efficacité)

            # Calcul du nombre de panneaux nécessaires
            puissance_panneau = 0.3  # kW (300 W)
            effectif_panneaux = int(puissance_requise / puissance_panneau) + 1  # Arrondi au nombre entier supérieur

            # Calcul de la surface nécessaire pour les panneaux
            surface_panneau = 1.6 * 1  # m²
            surface_necessaire = effectif_panneaux * surface_panneau

            # Exemple de logique de recommandation
            if surface < 10 and consommation < 200:
                recommendation = f"Panneaux PV de petite taille (5 kW) - {effectif_panneaux} panneaux de 1.6 m x 1 m"
            elif surface < 20 and consommation < 500:
                recommendation = f"Panneaux PV de taille moyenne (10 kW) - {effectif_panneaux} panneaux de 1.6 m x 1 m"
            else:
                recommendation = f"Panneaux PV de grande taille (20 kW) - {effectif_panneaux} panneaux de 1.6 m x 1 m"

            if budget < 5000:
                recommendation += "\nAttention: Vérifiez votre budget pour cette installation."
            if surface_necessaire > surface:
                recommendation += "\nAttention: La surface nécessaire dépasse la surface disponible."

            self.recommendation_text.delete('1.0', tk.END)
            self.recommendation_text.insert(tk.END, recommendation)

        except ValueError:
            messagebox.showerror("Erreur", "Veuillez saisir des valeurs numériques.")

    def run(self):
        self.window.mainloop()

if __name__ == "__main__":
    app = PVRecommender()
    app.run()
```
